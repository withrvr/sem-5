```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Car Data.csv")
fig=plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(df['Mileage'],bins = 10)
plt.title('Car Brand')
plt.xlabel('Mileage')
plt.ylabel('Pata nahi')
plt.show()
```

```python
fig=plt.figure()

ax = fig.add_subplot(1,1,1)
ax.scatter(df['Mileage'],df['Price'])
plt.title('Mileage and Price distribution')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```
